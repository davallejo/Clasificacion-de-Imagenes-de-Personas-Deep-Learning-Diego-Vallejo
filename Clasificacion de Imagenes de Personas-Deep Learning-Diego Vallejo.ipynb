```python
import os
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, precision_recall_curve
import tensorflow as tf
from tensorflow.keras import layers, models

# Definir rutas de las carpetas
base_dir = "FaceMaskDataset"
train_dir = os.path.join(base_dir, "Train")
test_dir = os.path.join(base_dir, "Test")
val_dir = os.path.join(base_dir, "Validation")

# Rutas de las carpetas de entrenamiento
train_with_mask_dir = os.path.join(train_dir, "WithMask")
train_without_mask_dir = os.path.join(train_dir, "WithoutMask")

# Rutas de las carpetas de prueba
test_with_mask_dir = os.path.join(test_dir, "WithMask")
test_without_mask_dir = os.path.join(test_dir, "WithoutMask")

# Rutas de las carpetas de validación
val_with_mask_dir = os.path.join(val_dir, "WithMask")
val_without_mask_dir = os.path.join(val_dir, "WithoutMask")

# Fase 1: Comprender el negocio (Business Understanding)

# En vista de las recientes crisis de salud que han afectado a nivel mundial en los últimos años, el uso generalizado de máscaras se ha convertido en una medida común tanto para protegerse a uno mismo como para proteger a los demás. Sin embargo, esta práctica también ha dado lugar a preocupaciones de seguridad, ya que puede dificultar la identificación y el reconocimiento de las personas en entornos de seguridad. Por consiguiente, como medida inicial para actualizar y fortalecer el sistema de seguridad de nuestra empresa, se nos ha encomendado la tarea de desarrollar e implementar un sistema capaz de identificar con precisión si una persona lleva puesta una máscara o no. Para abordar esta tarea, contamos con una extensa base de datos de imágenes que servirán como referencia para los conjuntos de entrenamiento, prueba y validación. 
# El objetivo principal de este proyecto es diseñar y entrenar un modelo de red neuronal que pueda distinguir y asignar etiquetas a las diferentes categorías de imágenes, proporcionando así una solución tecnológica efectiva para mejorar la seguridad en nuestros entornos.


# Fase 2: Comprensión de los datos (Data Understanding)

# Función para mostrar las imágenes de una carpeta
def show_images(directory, title):
    plt.figure(figsize=(15, 10))
    plt.suptitle(title, fontsize=16)
    
    images = os.listdir(directory)[:5]  # Obtener las primeras 5 imágenes
    for i, image_name in enumerate(images, 1):
        image_path = os.path.join(directory, image_name)
        
        # Cargar la imagen
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB
        
        plt.subplot(2, 5, i)
        plt.imshow(image)  # Mostrar la imagen
        plt.axis('off')
    
    plt.show()

# Mostrar las imágenes de cada carpeta
show_images(train_with_mask_dir, "Training Images with Mask")
show_images(train_without_mask_dir, "Training Images without Mask")
show_images(test_with_mask_dir, "Test Images with Mask")
show_images(test_without_mask_dir, "Test Images without Mask")
show_images(val_with_mask_dir, "Validation Images with Mask")
show_images(val_without_mask_dir, "Validation Images without Mask")

# Función para obtener estadísticas de las imágenes en una carpeta
def image_statistics(directory):
    image_data = []
    for img_name in os.listdir(directory):
        img_path = os.path.join(directory, img_name)
        img = cv2.imread(img_path)

        # Obtener dimensiones de la imagen
        height, width, _ = img.shape

        # Calcular el tamaño en disco (en bytes)
        size_bytes = os.path.getsize(img_path)

        # Obtener número de píxeles
        num_pixels = height * width

        image_data.append({
            'Image': img_name,
            'Width': width,
            'Height': height,
            'Pixels': num_pixels,
            'Size (Bytes)': size_bytes
        })

    # Convertir los datos a DataFrame de pandas
    df = pd.DataFrame(image_data)
    return df

# Función para visualizar estadísticas y gráficos de distribución
def visualize_statistics(df, title):
    print(f"\n{title}:")
    print(df.describe())

    plt.figure(figsize=(12, 6))
    plt.subplot(2, 2, 1)
    sns.histplot(df['Width'], kde=True, color='skyblue')
    plt.title('Distribution of Width')

    plt.subplot(2, 2, 2)
    sns.histplot(df['Height'], kde=True, color='salmon')
    plt.title('Distribution of Height')

    plt.subplot(2, 2, 3)
    sns.histplot(df['Pixels'], kde=True, color='green')
    plt.title('Distribution of Pixels')

    plt.subplot(2, 2, 4)
    sns.histplot(df['Size (Bytes)'], kde=True, color='orange')
    plt.title('Distribution of Size (Bytes)')

    plt.tight_layout()
    plt.show()

# Obtener estadísticas y visualizar para las imágenes con máscara de entrenamiento
df_train_with_mask = image_statistics(train_with_mask_dir)
visualize_statistics(df_train_with_mask, "Statistics for Training Images with Mask")

# Obtener estadísticas y visualizar para las imágenes sin máscara de entrenamiento
df_train_without_mask = image_statistics(train_without_mask_dir)
visualize_statistics(df_train_without_mask, "Statistics for Training Images without Mask")

# Obtener estadísticas y visualizar para las imágenes con máscara de prueba
df_test_with_mask = image_statistics(test_with_mask_dir)
visualize_statistics(df_test_with_mask, "Statistics for Test Images with Mask")

# Obtener estadísticas y visualizar para las imágenes sin máscara de prueba
df_test_without_mask = image_statistics(test_without_mask_dir)
visualize_statistics(df_test_without_mask, "Statistics for Test Images without Mask")

# Obtener estadísticas y visualizar para las imágenes con máscara de validación
df_val_with_mask = image_statistics(val_with_mask_dir)
visualize_statistics(df_val_with_mask, "Statistics for Validation Images with Mask")

# Obtener estadísticas y visualizar para las imágenes sin máscara de validación
df_val_without_mask = image_statistics(val_without_mask_dir)
visualize_statistics(df_val_without_mask, "Statistics for Validation Images without Mask")

# Fase 3: Preparación de los datos (Data Preparation)
# Función para cargar las imágenes y etiquetas
def load_data(directory):
    X = []
    y = []
    for folder in os.listdir(directory):
        label = 0 if folder == 'WithoutMask' else 1
        folder_path = os.path.join(directory, folder)
        for img in os.listdir(folder_path):
            img_path = os.path.join(folder_path, img)
            image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Convertir a escala de grises
            image = cv2.resize(image, (150, 150))  # Redimensionar a 150x150
            X.append(image)
            y.append(label)
    return np.array(X), np.array(y)

# Cargar datos de entrenamiento, prueba y validación
X_train, y_train = load_data(train_dir)
X_test, y_test = load_data(test_dir)
X_val, y_val = load_data(val_dir)

# Normalizar los valores de píxeles
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
X_val = X_val.astype('float32') / 255.0

# Fase 4: Modelado (Modeling)
# Modelo
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

# Compilar el modelo
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
               loss='binary_crossentropy',
               metrics=['accuracy'])

# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=6, batch_size=64, validation_data=(X_val, y_val))

# Fase 5: Evaluación (Evaluation)
# Matriz de confusión y métricas para el modelo
print("\nModelo:")
# Matriz de confusión
y_pred = (model.predict(X_test) > 0.5).astype("int32")
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualizar la matriz de confusión
classes = ['Without Mask', 'With Mask']
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=classes)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# Métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")

# Curva ROC y AUC-ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Curva PR y AUC-PR
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
pr_auc = auc(recall, precision)

plt.figure()
plt.plot(recall, precision, color='blue', lw=2, label='PR curve (area = %0.2f)' % pr_auc)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.show()

# Obtener historial de entrenamiento
train_loss = history.history['loss']
train_acc = history.history['accuracy']
val_loss = history.history['val_loss']
val_acc = history.history['val_accuracy']

# Graficar curvas de pérdida
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, len(train_loss) + 1), train_loss, 'b-', label='Training Loss')
plt.plot(range(1, len(val_loss) + 1), val_loss, 'r-', label='Validation Loss')
plt.title('Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Graficar curvas de precisión
plt.subplot(1, 2, 2)
plt.plot(range(1, len(train_acc) + 1), train_acc, 'b-', label='Training Accuracy')
plt.plot(range(1, len(val_acc) + 1), val_acc, 'r-', label='Validation Accuracy')
plt.title('Accuracy vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# Fase 6: Despliegue (Deployment)
# Guardar el modelo entrenado
model.save("modelo_entrenado.h5")
```

# Uso del modelo entrenado para prededir

```python
#Fase 6 Despliegue (Deployment)
import os
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Cargar el modelo entrenado
modelo_entrenado = load_model("modelo_entrenado.h5")

# Definir la ruta de la carpeta de validación con y sin máscara
val_with_mask_dir = "FaceMaskDataset/Validation/WithMask"
val_without_mask_dir = "FaceMaskDataset/Validation/WithoutMask"

# Obtener la lista de archivos de imágenes con y sin máscara de la carpeta de validación
val_with_mask_files = os.listdir(val_with_mask_dir)
val_without_mask_files = os.listdir(val_without_mask_dir)

# Ejecutar la predicción 5 veces
for _ in range(5):
    # Elegir aleatoriamente una imagen con máscara y otra sin máscara de la carpeta de validación
    imagen_con_mascara_file = random.choice(val_with_mask_files)
    imagen_sin_mascara_file = random.choice(val_without_mask_files)

    # Leer las imágenes
    imagen_con_mascara = cv2.imread(os.path.join(val_with_mask_dir, imagen_con_mascara_file), cv2.IMREAD_GRAYSCALE)
    imagen_sin_mascara = cv2.imread(os.path.join(val_without_mask_dir, imagen_sin_mascara_file), cv2.IMREAD_GRAYSCALE)

    # Preprocesar las imágenes (normalizar y redimensionar)
    imagen_con_mascara = cv2.resize(imagen_con_mascara, (150, 150)) / 255.0
    imagen_sin_mascara = cv2.resize(imagen_sin_mascara, (150, 150)) / 255.0

    # Agregar una dimensión adicional para la entrada del modelo
    imagen_con_mascara = np.expand_dims(imagen_con_mascara, axis=-1)
    imagen_sin_mascara = np.expand_dims(imagen_sin_mascara, axis=-1)

    # Realizar la predicción para la imagen con máscara
    prediccion_con_mascara = modelo_entrenado.predict(np.array([imagen_con_mascara]))
    resultado_con_mascara = "con máscara" if prediccion_con_mascara > 0.5 else "sin máscara"

    # Realizar la predicción para la imagen sin máscara
    prediccion_sin_mascara = modelo_entrenado.predict(np.array([imagen_sin_mascara]))
    resultado_sin_mascara = "con máscara" if prediccion_sin_mascara > 0.5 else "sin máscara"

    # Mostrar las imágenes y los resultados de predicción
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(imagen_con_mascara, cmap='gray')
    plt.title('Imagen con Máscara')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(imagen_sin_mascara, cmap='gray')
    plt.title('Imagen sin Máscara')
    plt.axis('off')

    plt.show()

    print(f"Predicción para la imagen con máscara: {resultado_con_mascara}")
    print(f"Predicción para la imagen sin máscara: {resultado_sin_mascara}")
```
